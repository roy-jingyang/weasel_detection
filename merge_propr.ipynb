# Merge to anonymize case ids in the proprietary dataset

```python
fn_logs = {
    'propr': './data/proprietary.csv',
    'propr_anonymized': './data/proprietary1.csv',
}
```

```python
import pandas as pd
```

```python
df = pd.read_csv(fn_logs['propr'])
df = df.dropna(how='all')
print(df['PI_ID'].nunique())
df
```

```python
df_new = pd.read_csv(fn_logs['propr_anonymized'])
df_new = df_new.dropna(how='all')
print(df_new['PI_ID'].nunique())
df_new
```

```python
df_new = df_new.merge(
    df, how='inner', 
    on=[
        'Activity', 'Start', 'End', 'Originator', 'Weekday', 'Cost',
    ],
    suffixes=['_anonymized', '_original']
)
mapping = df_new[['PI_ID_anonymized', 'PI_ID_original']].drop_duplicates()
print(len(mapping))
gb_anonymized = mapping.groupby('PI_ID_anonymized').size().to_frame().reset_index().rename(columns={0: 'size'})
gb_original = mapping.groupby('PI_ID_original').size().to_frame().reset_index().rename(columns={0: 'size'})
# OUTPUT: 12 rows (i.e., 12 PI_ID_anonymized values are mapped to more than one PI_ID_original)
gb_anonymized[gb_anonymized['size'] > 1]
# OUTPUT: 12 rows (i.e., 12 PI_ID_original values are mapped to more than one PI_ID_anonymized)
# gb_original[gb_original['size'] > 1]
```
