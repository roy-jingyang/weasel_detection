# Activity Deviation

## Description
Activities are observed in the log that are not a part of the process model.

```python
import pandas as pd
import numpy as np
import altair as alt
alt.data_transformers.enable('vegafusion')
```

```python
ref_log_name = 'bpic15_ref'
test_log_name = 'bpic15_test'

fn_logs = {
    'bpic15': '../data/preproc/bpic15.csv',
    'bpic15_ref': '../data/preproc/bpic15_train-70.csv',
    'bpic15_test': '../data/preproc/bpic15_test-30.csv',

    'bpic17': '../data/preproc/bpic2017.csv',

    'propr': '../data/preproc/proprietary.csv',
    'propr_ref': '../data/preproc/proprietary_train-70.csv',
    'propr_test': '../data/preproc/proprietary_test-30.csv'
}
```

```python
log_ref = pd.read_csv(fn_logs[ref_log_name])
print(log_ref['case_id'].nunique())
log_ref
```

```python
log_test = pd.read_csv(fn_logs[test_log_name])
print(log_test['case_id'].nunique())
log_test
```

## Rerouting-AD-1

### Extract two sets of activity labels

```python
s_ref_act = set(log_ref['activity'].unique())
s_test_act = set(log_test['activity'].unique())
```

### Calculate log coverage (how much of the log is in the model?)

```python
print('Log coverage =\t{}'.format(
    len(s_ref_act.intersection(s_test_act)) /
    len(s_test_act)
))
```

```python
print('The following activities are observed in the log, but are not in the reference:')
print('-' * 80)
for a in sorted(s_test_act.difference(s_ref_act)):
    print(a)
print('-' * 80)
print('The following events record these Activity Deviation instances:')
print('-' * 80)
log_test.loc[log_test['activity'].isin(s_test_act.difference(s_ref_act))]
```

## Rerouting-AD-3

### Calculate the relative frequency for each activity label

```python
rel_act_freq = (log_test.groupby('activity').size() / len(log_test)).reset_index().rename(columns={0: 'frequency'})
```

### Compare and find activity labels that are statistically significantly infrequent

```python
iqr = (
    np.quantile(rel_act_freq['frequency'], q=0.75) - 
    np.quantile(rel_act_freq['frequency'], q=0.25)
)
# print(iqr)

ub = np.quantile(rel_act_freq['frequency'], q=0.75) + 1.5 * iqr
lb = np.quantile(rel_act_freq['frequency'], q=0.25) - 1.5 * iqr

rel_act_freq['is_infrequent'] = rel_act_freq['frequency'].apply(
    lambda x: x < lb
)

alt.Chart(rel_act_freq).mark_boxplot().encode(
    y='frequency:Q'
)
```

```python
alt.Chart(rel_act_freq).mark_bar().encode(
    x='activity:O',
    y='frequency:Q',
    color='is_infrequent:N'
)
```
