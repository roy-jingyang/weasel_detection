# Originator Deviation

## Description
Activities are undertaken by resources that are not (officially) allowed to work on them.

```python
import pandas as pd
import numpy as np
import altair as alt
alt.data_transformers.enable('vegafusion')
```

```python
ref_log_name = 'propr_ref'
test_log_name = 'propr_test'

fn_logs = {
    'propr': '../data/preproc/proprietary.csv',
    'propr_ref': '../data/preproc/propr_train-70.csv',
    'propr_test': '../data/preproc/propr_test-30.csv'
}
```

```python
log_ref = pd.read_csv(fn_logs[ref_log_name])
print(log_ref['case_id'].nunique())
log_ref
```

```python
log_test = pd.read_csv(fn_logs[test_log_name])
print(log_test['case_id'].nunique())
log_test
```

## Rerouting-OD-1

```python
# construct the LTL rules from the reference model
rules = dict()
for event in log_ref.itertuples():
    act, res = getattr(event, 'activity'), getattr(event, 'resource')
    if act in rules:
        rules[act].add(res)
    else:
        rules[act] = set()
print(rules)
print('There are {} rules'.format(len(rules)))
```

### Perform rule checking to check compliance

```python
deviated_events_idx = list()
deviated_act_res = dict()
violated_rules = set()
for index, event in log_test.iterrows():
    act = getattr(event, 'activity')
    res = getattr(event, 'resource')
    if act in rules and len(rules[act]) > 0:
        # do checking if rules exist
        if res in rules[act]:
            # complied
            pass
        else:
            # non-complied
            deviated_events_idx.append(index)
            if (act, res) in deviated_act_res:
                deviated_act_res[(act, res)] += 1
            else:
                deviated_act_res[(act, res)] = 1
            violated_rules.add(act)
    else:
        # do nothing as no rule exists
        pass

print('The following activities were undertaken by resources that are not (officially) allowed to work on them, according to the reference:')
print('-' * 80)
for ar, count in deviated_act_res.items():
    a, r = ar
    print('Activity "{}" performed by Resource "{}" {} times'.format(
        a, r, count
    ), end= '\t')
    print('however the resource is not among the {} allowed resources.'.format(
        len(rules[a])
    ))
print('-' * 80)

print('The following {} rules are violated: {}'.format(
    len(violated_rules), sorted(violated_rules)
))

print('The following events in the log record Originator Deviation:')
print('These events are recorded in {} cases'.format(log_test.loc[deviated_events_idx, 'case_id'].nunique()))
print('These events were performed by {} unique resources'.format(log_test.loc[deviated_events_idx, 'resource'].nunique()))
print('-' * 80)
log_test.loc[deviated_events_idx]
```
