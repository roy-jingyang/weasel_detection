```python
import pandas as pd

# load datasets
df_old = pd.read_excel('./data/Dataset_Original.xlsx')
df_new = pd.read_excel('./data/Dataset_Anonymized.xlsx')
```

```python
df_new
```

```python
# use only common columns
df_old = df_old[df_new.columns]
df_old
```

```python
# assuming line indices match, check if cell values match 

# compare() returns a DataFrame showing the differences
# if return empty, then the two have equal values
# EXPECT: only the two PI_ID columns are unmatched (due to anonymization)
df_diff = df_old.compare(df_new).rename(columns={'self': 'original', 'other': 'anonymized'})
df_diff

# TRUE: indices match
```

```python
# ignore the extra 'PI_ID' name in the column
df_diff.columns = df_diff.columns.get_level_values(1)
df_diff
```

```python
# check the mapping
# EXPECT: one-to-one mapping between unique values of `original` and
# `anonymized`

df_diff.drop_duplicates()
```

```python
df_old['PI_ID_anonymized'] = df_new['PI_ID']
df_old
```

```python
df_old[['PI_ID', 'PI_ID_anonymized']].groupby('PI_ID').agg(
    num_anonymized_values=pd.NamedAgg('PI_ID_anonymized', 'nunique')
)['num_anonymized_values'].unique()
```

```python
df_old[['PI_ID', 'PI_ID_anonymized']].groupby('PI_ID_anonymized').agg(
    num_original_values=pd.NamedAgg('PI_ID', 'nunique')
)['num_original_values'].unique()
```
